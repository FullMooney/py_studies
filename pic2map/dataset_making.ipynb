```python
import cv2
```

```python
w = 128
h = 128
w_start = 80
w_end = 1872
h_start = 80
h_end = 976

# dst = src.copy()
for x in range(4):
    srcA = cv2.imread("images/dataset/A{}.png".format(x), cv2.IMREAD_COLOR)
    srcB = cv2.imread("images/dataset/B{}.png".format(x), cv2.IMREAD_COLOR)
    for i in range((h_end-h_start)//h):
        for j in range((w_end-w_start)//w):
            tempA = srcA[h_start + i*h : h_start + (i+1)*h, w_start + j*w : w_start + (j+1)*w]
            print('images/trainA/A{}_{}_{}.jpg'.format(x, str(i).zfill(2), str(j).zfill(2)))
            cv2.imwrite('images/trainA/A{}_{}_{}.jpg'.format(x, str(i).zfill(2), str(j).zfill(2)), tempA)
            tempB = srcB[h_start + i*h : h_start + (i+1)*h, w_start + j*w : w_start + (j+1)*w]
            print('images/trainB/B{}_{}_{}.jpg'.format(x, str(i).zfill(2), str(j).zfill(2)))
            cv2.imwrite('images/trainB/B{}_{}_{}.jpg'.format(x, str(i).zfill(2), str(j).zfill(2)), tempB)
            

# cv2.imshow("src", src)

# cv2.destroyAllWindows()
```
